```python
import json
import pandas as pd
```

```python
ocean_data = open("ocean-snapshot-vote-data.json","r")
```

```python
data = json.load(ocean_data)
```

```python
ocean_df = pd.json_normalize(data['data'], record_path = ['votes'])
```

```python
ocean_df.dropna(subset = ["choice"], inplace = True)
```

```python
ocean_df.head()
```

```python
cols_to_drop = ["choice." + str(k) for k in range(1,40)]
```

```python
ocean_df_cols = list(ocean_df)
for k in range(1,43):
    choice_num = "choice" + "." + str(k)
    if choice_num in ocean_df_cols:
        ocean_df.drop(columns = [choice_num], inplace = True)

ocean_df.drop(columns = ["id"], inplace = True)
```
