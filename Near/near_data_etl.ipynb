## Near Telegram Data Extract Transform Load

Get a list of all files.

```python
import glob

json_files = glob.glob("./data/Telegram Data/*")
json_files
```

Investigate a single file.

```python
import json

with open(json_files[0]) as file:
    data = json.load(file)
data.keys()
```

```python
import pandas as pd
pd.DataFrame(data)
```

Load the messages data as a dataframe.

```python
df = pd.DataFrame(data['messages'])
df['data_name'] = data['name']
df['data_type'] = data['type']
df['data_id'] = data['id']
df
```

Looks like we have some NaN. Let's check.

```python
df.isna().sum() / len(df)
```

Let's check all the files.

```python
data_frames = []
for json_file in json_files:
    with open(json_file) as file:
        data = json.load(file)
    df = pd.DataFrame(data['messages'])
    df['data_name'] = data['name']
    df['data_type'] = data['type']
    df['data_id'] = data['id']
    data_frames.append(df)
```

```python
all_data = pd.concat(data_frames)
```

```python
all_data
```

```python
all_data.isna().sum() / len(all_data)
```

Drop columns with more than 90% NaN

```python
data_90 = all_data[['id', 'type', 'date', 'text', 'from', 'from_id', 'reply_to_message_id', 'data_name', 'data_type', 'data_id']]
data_90
```

Drop columns with more than 98% NaN

```python
data_98 = all_data[['id', 'type', 'date', 'actor', 'actor_id', 'text', 'from', 'from_id', 'reply_to_message_id', 'edited', 'members', 'file', 'media_type', 'width', 'height', 'thumbnail', 'sticker_emoji', 'photo',  'data_name', 'data_type', 'data_id']]
data_98
```

```python
data_16 = data_98.dropna(thresh=16)
data_16
```

```python
data_16.isna().sum() / len(data_16)
```

```python
some_clean_telegram_data = data_16.drop(['actor', 'actor_id', 'edited', 'members', 'photo'], axis=1)
some_clean_telegram_data
```

```python
some_clean_telegram_data.to_csv('outputs/some_clean_telegram_data.csv', index=False)
```
